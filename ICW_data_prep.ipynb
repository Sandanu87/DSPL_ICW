```python
def list_different_columns(df_list):
    all_columns = set()
    for df in df_list:
        all_columns.update(df.columns)

    common_columns = set(df_list[0].columns)
    for df in df_list[1:]:
        common_columns.intersection_update(df.columns)

    different_columns = list(all_columns - common_columns)
    return different_columns

df_list = [df_10, df_11, df_12]
different_cols = list_different_columns(df_list)

if different_cols:
    print("Different columns in the DataFrames:")
    for col in different_cols:
        print(col)
else:
    print("All DataFrames have the same columns.")
```

```python
column_mapping = {
    'Crime ategory': 'Crime Category',
    'Galle (Galle/ Elpitiya)': 'Galle (Galle/Elpitiya)',
    'Badulla (Badfulla & Bandarawela )': 'Badulla (Badulla & Bandarawela )',}

def replace_different_columns(df_list, column_mapping):

    modified_df_list = []
    for df in df_list:
        modified_df = df.rename(columns=column_mapping, errors='ignore')
        modified_df_list.append(modified_df)
    return modified_df_list

modified_df_list = replace_different_columns(df_list, column_mapping)
df_10, df_11, df_12 = modified_df_list
```

```python
def melt_crime_data(df):
    district_columns = [col for col in df.columns if col != 'Crime Category']
    melted_data = df.melt(id_vars=['Crime Category'],
                         value_vars=district_columns,
                         var_name='District',
                         value_name='Value')
    return melted_data

melted_df_10 = melt_crime_data(df_10)
melted_df_11 = melt_crime_data(df_11)
melted_df_12 = melt_crime_data(df_12)

crime_data = pd.concat([melted_df_10, melted_df_11, melted_df_12], ignore_index=True)
```

```python
merged_crime_data = pd.merge(melted_df_10, melted_df_11, on=['District', 'Crime Category'], how='inner')
merged_crime_data = pd.merge(merged_crime_data, melted_df_12, on=['District', 'Crime Category'], how='inner')

merged_crime_data.rename(columns={'Value_x': '2010', 'Value_y': '2011', 'Value': '2012'}, inplace=True)
```
